```python
import numpy as np
import matplotlib.pyplot as plt


p = 0.4 # probabilité de succès
N = 10 # taille de l'échantillon

# Nombre de simulations
num_simulations = 300

echantillons = np.random.random((num_simulations, N)) < p
resultats = np.sum(echantillons, axis=1)

plt.figure(figsize=(12, 8))
plt.hist(resultats, bins=np.arange(N+2)-0.5, edgecolor='black', alpha=0.7)

# Ajout de labels et titre
plt.xlabel('Nombre de bactéries a1')
plt.ylabel('Nombre d\'échantillons')
plt.title('Distribution expérimentale du nombre de bactéries a1 dans un échantillon de taille 10')

plt.xticks(np.arange(N+1))  # Ajuster les marques sur l'axe x

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_evolution(initial_frequency, num_generations, population_size):

    # Init
    frequencies = [initial_frequency]

    for generation in range(1, num_generations + 1):
        
        frequency_at_generation = np.random.binomial(population_size, frequencies[generation-1]) / population_size
        frequencies.append(frequency_at_generation)

    return frequencies


initial_frequency = 0.4  # Fréquence initiale de l'allèle a1
num_generations = 10  # Nombre de générations
population_size = 10  # Taille de la population

frequencies = simulate_evolution(initial_frequency, num_generations, population_size)

generations = list(range(num_generations + 1))
plt.figure(figsize=(12, 8))
plt.plot(generations, frequencies, marker='o')
frame_text = f'Taille de population initiale : {population_size}'
plt.annotate(frame_text, xy=(0.05, 0.92), xycoords='axes fraction', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.xlabel('Générations')
plt.ylabel('Fréquence de l\'allèle a1')
plt.title('Évolution de la fréquence de l\'allèle a1 sur les générations')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_evolution(initial_frequency, num_generations, population_size):
    frequencies = [initial_frequency]

    for generation in range(1, num_generations + 1):
        # Si nous atteignons la moitié des simulations, réduire la population
        if generation == num_generations // 2:
            population_size //= 2  # Réduire la population de moitié

        frequency_at_generation = np.random.binomial(population_size, frequencies[generation-1]) / population_size
        frequencies.append(frequency_at_generation)

    return frequencies

initial_frequency = 0.4
num_generations = 100
population_size = 100

frequencies = simulate_evolution(initial_frequency, num_generations, population_size)

plt.figure(figsize=(10, 6))

generations = list(range(num_generations + 1))
plt.plot(generations, frequencies, marker='o')
plt.axvline(x=num_generations // 2, color='r', linestyle='--', label='Réduction pop.')  # Ligne verticale rouge
frame_text = f'Taille de population initiale : {population_size}'
plt.annotate(frame_text, xy=(0.05, 0.92), fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.xlabel('Générations')
plt.ylabel('Fréquence de l\'allèle a1')
plt.title('Évolution de la fréquence de l\'allèle a1 sur les générations')
plt.legend()

plt.show()

```
